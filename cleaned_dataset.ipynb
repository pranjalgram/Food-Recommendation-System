```python
def reduce_na(cal):
    store = ""
    save = cal.split(';')
    for i in save:
        if i=='Protein':
            store=store+i+';'
        if i=='Carbohydrates':
            store=store+i+';'
        if i=='Fats':
            store=store+i+';'
        if i=='Calcium':
            store=store+i+';'
        if i=='Vitamins':
            store=store+i+';'     
    if store!="":
        if store[-1]==';':
            cal = store[:-1]
    else:
        cal = 'Protein'
    return cal

print(reduce_na('Protein;Vitamins;Taste too ðŸ˜›'))
```
